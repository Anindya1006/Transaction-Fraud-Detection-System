```python
# feature engineering and training ml model
# used to split the dataset into train and test split
from sklearn.model_selection import train_test_split
# data transformation
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# training the model and transformation operations together
from sklearn.pipeline import Pipeline
# XGBoost model
from xgboost import XGBClassifier
# for evaluation of the model
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
df.head()
```

```python
df_model=df.drop(["nameOrig","nameDest","isFlaggedFraud"],axis=1)
```

```python
df_model.head()
```

```python
categorical=["type"]
numeric=["amount","oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest","balanceDiffOrig","balanceDiffDest"]
```

```python
y=df_model["isFraud"]
X=df_model.drop("isFraud",axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y,random_state=42)
```

```python
preprocessor = ColumnTransformer(
transformers= [
    ("num","passthrough",numeric),
    ("cat",OneHotEncoder(drop="first",handle_unknown="ignore"),categorical)
],
  remainder="drop"
)
```

```python
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

pipeline=Pipeline([
    ("prep",preprocessor),
    ("clf",XGBClassifier(
        n_estimators=200,
        max_depth=6,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="binary:logistic",
        eval_metric="logloss",
        scale_pos_weight=scale_pos_weight,
        tree_method="hist",
        random_state=42,
        n_jobs=-1
    ))
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred=pipeline.predict(X_test)
y_pred_proba=pipeline.predict_proba(X_test)[:,1]
```

```python
print(classification_report(y_test,y_pred))
print("ROC-AUC:", roc_auc_score(y_test,y_pred_proba))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pipeline.score(X_test,y_test)*100
```
